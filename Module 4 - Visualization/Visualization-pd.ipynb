```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(data = np.random.rand(100,2), columns=list('AB'))
df['C'] = 2*df['A']*df['A'] + 3*df['A'] + 5
df['D'] = df['A'] + df['B']
df['E'] = df['A']>df['B']
df.head()
```

```python
# line plot between x=A and y=C
# sort by A first
df.sort_values(by=['A']).plot(x='A', y='C')
```

```python
# scatter plot between x=A and y=C
df.plot(x='A', y='C', kind='scatter')
```

```python
# bar plot of number of  instances of A where A>B
df.groupby('E')['A'].count().plot(kind='bar')
```

```python
# pie plot of number of  instances of A where A>B
df.groupby('E')['A'].count().plot(kind='pie')
```

```python
# histogram of A using 10 bins
df['A'].plot.hist(bins=10)
```

```python
# box plot
df.plot.box() # df['A'].plot.box()
```

```python
# area plot
df.plot.area() 
```
