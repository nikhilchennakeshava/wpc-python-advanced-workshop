```python
##### Seaborn
#
# - Seaborn is a library for making statistical graphs in Python. 
# - It is built on top of matplotlib and closely integrates with pandas data structures.
# - Two way to access sns functionality:
#       Figure-level interfaces provide conveniet wrappers for multiple underlying axes-level functions; these return FacetGrid objects
#       Axes-level functions provide more control when combining multiple figures into a single figure; these return axes objects
#
# https://seaborn.pydata.org/
# https://seaborn.pydata.org/introduction.html
# https://seaborn.pydata.org/examples/index.html
#
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
##### setting and checking color palette: deep, muted, pastel, bright, dark, and colorblind
sns.set_palette("deep")
sns.palplot(sns.color_palette()) 
```

```python
##### load sns sample data sets
# https://github.com/mwaskom/seaborn-data
tips = sns.load_dataset('tips')
tips.head()
```

```python
##### relational
# https://seaborn.pydata.org/tutorial/relational.html
# figure-level interface: sns.relplot()
#      encompasses axes-level functions: scatterplot() and lineplot()
```

```python
# scatter plot between 2 variables (relplot plots scatter by default, can change using kind)
sns.relplot(data=tips, x="total_bill", y="tip");
```

```python
# scatter plot between 2 variables, while segmenting based on a 3rd variable (using hue)
sns.relplot(data=tips, x="total_bill", y="tip", hue='smoker');
```

```python
# scatter plot between 2 variables, while segmenting based on a 3rd variable (using size)
sns.relplot(data=tips, x="total_bill", y="tip", size='size');
```

```python
# line plot between 2 variables: can sort data by x values
sns.relplot(data=tips, x="total_bill", y="tip", kind='line', sort=True); # try with sort=False
```

```python
# showing multiple relationships with facets: make multiple axes and plot subsets of the data on each of them
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="time", height=3, aspect=1);
```

```python
# showing multiple relationships with facets: make multiple axes and plot subsets of the data on each of them
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", row="sex", col="time", height=3, aspect=1);
```
